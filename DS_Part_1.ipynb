```python
###✅ NumPy Aggregate Functions
Aggregate functions are used to perform calculations
on array elements
such as sum, average, maximum, and minimum.
```

```python
import numpy as np

a = np.array([16, 25, 33, 40])

print("Maximum=", a.max())
print("Minimum=", a.min())
print("Sum=", a.sum())
print("Mean=", a.mean())
print("Standard deviation=", a.std())
print("Variance=", a.var())
```

```python
###✅ Indexing in NumPy
* Indexing is used to access specific elements in a NumPy array.
* It works using the position (index) of elements.
* Indexing can be done for both 1D and multidimensional arrays.
* In NumPy, indexing starts from 0.
```

```python
###✅ 1️⃣ Indexing in 1D Array
import numpy as np
a = np.array([10, 20, 30, 40, 50])
print("4th element=",a[3])
print("5th element=",a[-1])

###✅ 2️⃣ Indexing in 2D Array
b = np.array([[1, 2, 3],[4, 5, 6]])
print("row 2 column 2 element=",b[1,1])
print("row 1 column 3 element=",b[0,2])
```

```python
###✅ Slicing in NumPy
* Slicing is used to access a range of elements from an array.
* It allows selecting multiple elements at once.
* Syntax:
array[start:end]
```

```python
###✅ 1️⃣ Slicing in 1D Array
import numpy as np
a = np.array([10, 20, 30, 40, 50])
print(a[1:4:2])
print(a[:3])

###✅ 2️⃣ Slicing in 2D Array
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print(b[0:2, 1:3])
print(b[:,1]) ## print only second column
print(b[1,:]) ## print only 2nd row
```

```python
###✅ Reshaping in NumPy
* Reshaping means changing the shape or dimensions of a NumPy array without changing its data.
* It is done using the reshape() function.
* The total number of elements before and after reshaping must remain the same.
* Reshaping is used to convert arrays from one dimension to another, such as 1D to 2D or 2D to 3D.
* NumPy also provides:
    * flatten() → converts array into 1D array
    * ravel() → converts array into 1D array
```

```python
####✅ Example for Reshaping in NumPy

##Example 1: Reshaping 1D Array to 2D Array
import numpy as np
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3) #(rows nd columns)
print(b)

print("#####")
##Example 2: Reshaping into 3D Array
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(3, 1, 2) # 3 block, 1 row and 2 columns in each block
print(b)
```

```python
✅ What is a Series?

A Series is a one-dimensional labeled array in Pandas.

* Holds 1D data
* Has index and values
* Can store numbers, strings, dates, etc.
```

```python
### Example
import pandas as pd
s1 = pd.Series([10, 20, 30, 40])
s2 = pd.Series([90, 85, 88], index=['Math', 'Science', 'English'])

print(s1)
print(s2)
```

```python
✅ What is a DataFrame?

A DataFrame is a two-dimensional labeled data structure in Pandas.

* Contains rows and columns
* Similar to Excel sheet or SQL table
* Can store multiple data types
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [24, 30, 28],
    'Score': [85, 90, 88]
}
df = pd.DataFrame(data)
df.index=['a','b','c']
print(df)

###Access Columns
print("Access Columns")
print(df['Name'])

###Access rows
print("Access rows")
print(df.loc['b'])## using labelledindex
print(df.iloc[2])## using position

##Rename columns
print("Rename columns")
df.rename(columns={"Score": "Marks"}, inplace=True)

print(df)
```

```python
### Filtering, sorting, grouping
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [20, 21, 20, 21],
    "Score": [80, 90, 95, 100]
}
df = pd.DataFrame(data)
print(df)

#1---filtering
print("filtering")
print(df[df["Age"]>20])

# 2---Sorting
print("Sorting")
print(df.sort_values(by="Score",ascending=False))

#3---Grouping
##👉 “Group people by Age and find the average Score for each age.”
print("Grouping")
print(df.groupby("Age")["Score"].mean())
```

```python
###✅ Rank and Sort in Pandas

🔹 1️ Sorting in Pandas

✅ What is Sorting?
Sorting means arranging data in:
* Ascending order
* Descending order
* syntax: sort_values()

🔹 2️⃣ Ranking in Pandas
✅ What is Ranking?
Ranking assigns position numbers based on values.
* syntax: rank()

These are important operations used in:
* Data analysis
* Leaderboards
* Business reporting
* Top-k selection
```

```python
###✅ Example for Sorting
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Marks": [85, 95, 90]
}
df = pd.DataFrame(data)
print(df.sort_values(by="Marks"))
print("Descending")
print(df.sort_values(by="Marks", ascending=False))

```

```python
###✅ Example for Ranking
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie","lily"],
    "Marks": [85, 95, 90,90]
}
df = pd.DataFrame(data)
df["Rank1"]= df["Marks"]. rank() ## lower marks gets 1st rank
print(df)
df["Rank2"]= df["Marks"]. rank(ascending=False) ## highest marks get 1st rank
print(df)

##✅ Ranking Methods
#1---Dense Rank----No gaps in ranking.
df["Dense_Rank"] = df["Marks"].rank(method="dense", ascending=False)
print(df)

#2---Min Rank---Same minimum rank for ties.
df["Min_Rank"] = df["Marks"].rank(method="min", ascending=False)
print(df)

#3---Max Rank---Same maximum rank for ties.
df["Max_Rank"] = df["Marks"].rank(method="max", ascending=False)
print(df)

#4--First Rank---This gives ranks based on the order of appearance when values are tied.
df["First_rank"]=df["Marks"].rank(method="first", ascending=False)
print(df)
```
